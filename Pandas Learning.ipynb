```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('titanic.csv') #read csv file
```

```python
df.isnull().sum()
```

```python
df.describe().columns
```

```python
df_num = df[['Survived', 'Pclass', 'Age', 'Siblings/Spouses Aboard',
       'Parents/Children Aboard', 'Fare']]
```

```python
for i in df_num.columns:
    plt.hist(df_num[i])
    plt.title(i)
    plt.show()
```

```python
print(df_num.corr())
sns.heatmap(df_num.corr())
```
